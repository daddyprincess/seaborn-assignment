## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for 
creating informative and attractive statistical graphics. Here are five common types of plots that can be created using
Seaborn, along with their uses:

1.Scatter Plot:
    ~Use: Scatter plots are used to visualize the relationship between two continuous variables. They show individual data
     points as dots on a two-dimensional plane, making it easy to identify patterns, correlations, and outliers.
        
2.Histogram:
    ~Use: Histograms are used to visualize the distribution of a single continuous variable. They divide the data into bins 
     and display the frequency or count of values within each bin, allowing you to understand the data's underlying
    distribution.
    
3.Bar Plot (Categorical Plot):
    ~Use: Bar plots are used to display categorical data with rectangular bars of varying heights. They are useful for
     comparing different categories or groups based on a specific metric or count.
        
4.Box Plot (Box-and-Whisker Plot):
    ~Use: Box plots are used to visualize the distribution and summary statistics of a dataset. They display the median,
     quartiles, and potential outliers, making it easy to compare the spread and skewness of multiple datasets.
        
5.Heatmap:
    ~Use: Heatmaps are used to visualize the relationships and patterns in a dataset. They are often used for displaying
     correlations between variables in a matrix form, where colors represent the strength and direction of the correlation.
```

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.title("FMRI Signal Over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event", loc="upper right")
plt.show()
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create a figure with two subplots
plt.figure(figsize=(12, 6))

# Box plot for 'age' based on 'pclass'
plt.subplot(1, 2, 1)
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title("Box Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

# Box plot for 'fare' based on 'pclass'
plt.subplot(1, 2, 2)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title("Box Plot of Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram for the 'carat' column
plt.figure(figsize=(8, 6))
sns.histplot(data=diamonds_data, x="carat", bins=30, kde=True, color="skyblue")
plt.title("Histogram of Diamond Carat")
plt.xlabel("Carat")
plt.ylabel("Frequency")
plt.show()
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue based on the "species" column
sns.set(style="ticks")
sns.pairplot(iris_data, hue="species", palette="Set1")

# Display the pair plot
plt.show()
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to create a pivot table
flights_pivot = flights_data.pivot_table(index="month", columns="year", values="passengers")

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d", linewidths=0.5)
plt.title("Passenger Count by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
